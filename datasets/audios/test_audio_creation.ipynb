```python
from TTS.api import TTS
import os

MetadataFile = '/home/huitznahui/Documents/Code Experiments/Coqui/Tutorial 1/audios/metadata_tests_audios.csv'

TraditionalModelFolder = "/home/huitznahui/Documents/Code Experiments/Coqui/Tutorial 1/models/Marzo_GlowTTS_1.0.0/"
GeneratedModelFolder = "/home/huitznahui/Documents/Code Experiments/Coqui/Tutorial 1/models/Fuerza_Marzo_GlowTTS_1.1.1/"

TraditionalFolder = "/home/huitznahui/Documents/Code Experiments/Coqui/Tutorial 1/audios/Traditional/"
GeneratedFolder = "/home/huitznahui/Documents/Code Experiments/Coqui/Tutorial 1/audios/Generated/"
```

```python
def readMetadataFile(path:str):
    """
    Read the metadata file and return the lines as a list.
    """
    lines = []
    with open(path) as file:
        lines = file.readlines()
    return lines

def createAudioFiles(name:str, line:str, traditionalModel:TTS, generatedModel:TTS):
    """
    Create the audio files from the metadata file.
    """
    traditionalModel.tts_to_file(text=line,file_path=TraditionalFolder+name+".wav")
    generatedModel.tts_to_file(text=line,file_path=GeneratedFolder+name+".wav")

def main():
    """
    Main function.
    """
    # Read the metadata file
    lines = readMetadataFile(MetadataFile)

    # Create the models
    traditionalModel = TTS(model_path=TraditionalModelFolder+'best_model.pth', config_path=TraditionalModelFolder+'config.json')
    generatedModel = TTS(model_path=GeneratedModelFolder+'best_model.pth', config_path=GeneratedModelFolder+'config.json')

    # Create the audio files
    for line in lines:
        name = line.split("|")[0]
        text = line.split("|")[2]
        createAudioFiles(name=name, line=text, traditionalModel=traditionalModel, generatedModel= generatedModel)
    
```

```python
main()
```
